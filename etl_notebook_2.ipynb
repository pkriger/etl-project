```python
import pandas as pd
```

```python
names_df = pd.read_csv("Resources\IMDb names3.csv")
```

```python
movies_df = pd.read_csv("Resources\IMDb movies3.csv")
```

```python
title_principals_df = pd.read_csv("Resources\IMDb title_principals3.csv")
```

```python
join_df=pd.merge(names_df, title_principals_df, how='inner', on=['Name_ID'])
```

```python
join_df2= pd.merge(join_df, movies_df, how = 'inner', on= ['Title_ID'])
```

```python
join_df2.columns
```

```python
join_df2
data_df= join_df2[['Title_ID','Title','Year','Country', 'Genre',  'Director', 'Writer',
                   'Name_ID', 'Name',  'Category', 'Characters', 'Budget','Gross_Income']]
data_df
```

```python
data_df['Budget']=data_df['Budget'].str.strip('$')
data_df['Gross_Income']=data_df['Gross_Income'].str.strip('$')
#data_df['Genre']=data_df['Genre'].str.strip(',')
```

```python
data_df=data_df.replace(','," ",regex=True)
```

```python
data_df.to_csv("Resources\IMDb_dataframe.csv", index=False, header=True)
```
